```python
import pandas as pd
import matplotlib.pyplot as plt


faixa_etaria_sweden = pd.read_csv('sweden.csv', low_memory=False, index_col=0)
faixa_etaria_denmark = pd.read_csv('denmark.csv', low_memory=False, index_col=0)
faixa_etaria_norway = pd.read_csv('norway.csv', low_memory=False, index_col=0)
faixa_etaria_iceland = pd.read_csv('iceland.csv', low_memory=False, index_col=0)
faixa_etaria_finland = pd.read_csv('finland.csv', low_memory=False, index_col=0)

faixa_etaria_sweden['Masculino e Feminino'] = faixa_etaria_sweden['M'] + faixa_etaria_sweden['F']
faixa_etaria_denmark['Masculino e Feminino'] = faixa_etaria_denmark['M'] + faixa_etaria_denmark['F']
faixa_etaria_norway['Masculino e Feminino'] = faixa_etaria_norway['M'] + faixa_etaria_norway['F']
faixa_etaria_iceland['Masculino e Feminino'] = faixa_etaria_iceland['M'] + faixa_etaria_iceland['F']
faixa_etaria_finland['Masculino e Feminino'] = faixa_etaria_finland['M'] + faixa_etaria_finland['F']

faixa_etaria_sweden = faixa_etaria_sweden.drop(columns=['F', 'M'])
faixa_etaria_denmark = faixa_etaria_denmark.drop(columns=['F', 'M'])
faixa_etaria_norway = faixa_etaria_norway.drop(columns=['F', 'M'])
faixa_etaria_iceland = faixa_etaria_iceland.drop(columns=['F', 'M'])
faixa_etaria_finland = faixa_etaria_finland.drop(columns=['F', 'M'])


faixa_etaria_sweden.plot(kind = 'barh', color='darkblue', title='Faixa etria da populao na Sucia')
faixa_etaria_denmark.plot(kind = 'barh', color="olive", title='Faixa etria da populao na Dinamarca')
faixa_etaria_norway.plot(kind = 'barh', color="green", title='Faixa etria da populao na Noruega')
faixa_etaria_iceland.plot(kind = 'barh', color="lightcoral", title='Faixa etria da populao na Islndia')
faixa_etaria_finland.plot(kind = 'barh', color="slategrey", title='Faixa etria da populao na Finlndia')
```
