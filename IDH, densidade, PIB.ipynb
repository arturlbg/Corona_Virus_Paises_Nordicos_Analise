```python
import pandas as pd
import matplotlib.pyplot as plt

#SERIES PANDAS
serie_IDH = pd.Series({'Suecia':0.937, 'Dinamarca': 0.930, 'Noruega': 0.954, 'Finlandia':0.925, 'Islandia': 0.938}, name="IDH")
serie_Densidade = pd.Series({'Suecia':22.3, 'Dinamarca': 134.81, 'Noruega': 16.59, 'Finlandia': 16.38, 'Islandia': 3.34}, name="Densidade populacional")
serie_PIB = pd.Series({'Suecia':518.284, 'Dinamarca': 287.8, 'Noruega': 381.2, 'Finlandia': 244.9, 'Islandia':18.8}, name="PIB ")

#IDH
paises_idh = {'IDH': serie_IDH}
paises_idh = pd.DataFrame(paises_idh)
paises_idh.plot(kind = 'bar', title='Indice de Desenvolvimento Humano dos países, vai de 0 a 1')

#DENSIDADE POPULACIONAL
paises_densidade = {'Densidade populacional': serie_Densidade}
paises_densidade = pd.DataFrame(paises_densidade)
paises_densidade.plot(kind = 'bar', color="chocolate", title='Densidade populacional (população por km²)')

#PIB
paises_PIB = {'PIB': serie_PIB}
paises_PIB = pd.DataFrame(paises_PIB)
paises_PIB.plot(kind = 'bar', color="green", title='PIB por milhões de $')


```
