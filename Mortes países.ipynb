```python
import pandas as pd
import matplotlib.pyplot as plt

paises = pd.read_csv('countries.csv', low_memory=False, index_col=0)
paises_df = pd.DataFrame(paises)

sweden = paises_df.query('Country == "Sweden"')
sweden = sweden.drop(columns=['Confirmed', 'Recovered'])
sweden.index = pd.to_datetime(sweden.index)

denmark = paises_df.query('Country == "Denmark"')
denmark = denmark.drop(columns=['Confirmed', 'Recovered'])
denmark.index = pd.to_datetime(denmark.index)

norway = paises_df.query('Country == "Norway"')
norway = norway.drop(columns=['Confirmed', 'Recovered'])
norway.index = pd.to_datetime(norway.index)

finland = paises_df.query('Country == "Finland"')
finland = finland.drop(columns=['Confirmed', 'Recovered'])
finland.index = pd.to_datetime(finland.index)

iceland = paises_df.query('Country == "Iceland"')
iceland = iceland.drop(columns=['Confirmed', 'Recovered'])
iceland.index = pd.to_datetime(iceland.index)


fig, ax = plt.subplots()
ax.plot(sweden.index, sweden['Deaths'], label = 'Suécia')
ax.plot(denmark.index, denmark['Deaths'], label = 'Dinamarca')
ax.plot(norway.index, norway['Deaths'], label = 'Noruega')
ax.plot(finland.index, finland['Deaths'], label = 'Finlandia')
ax.plot(iceland.index, iceland['Deaths'], label = 'Islandia')
ax.set_xlabel('Data')
ax.set_ylabel('Mortes')
ax.set_title("Mortes nos países escandinavos por COVID-19")
ax.legend()
```
